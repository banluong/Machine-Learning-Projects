## K Means Cluster Creation

Now it is time to create the Cluster labels

```python
# Import KMeans for sklearn
from sklearn.cluster import KMeans
```

```python
# Instance with KMeans model, cluster size set as 2, private and non-private
kmeans = KMeans(n_clusters=2)
```

```python
# Fit model to all date, drop Private label
kmeans.fit(df.drop('Private',axis=1))
```

```python
# Cluster center vectors
kmeans.cluster_centers_
```

## Evaluation

Because there are labels for this dataset, we'll use this to our advantage to evaluate the clusters. Note in the real world we won't have labels.

```python
# Using apply and lambda to create new Cluster column, 1 for Private college, 0 for non-Private college
df['Cluster'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Check for new Cluster column
df.head()
```

```python
# Tools for evaluation from sklearn to see how well the Kmeans clustering worked without any labels
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
# labels
kmeans.labels_
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print('\n')
print(classification_report(df['Cluster'],kmeans.labels_))
print('\n')
print('Accuracy Score:', accuracy_score(df['Cluster'],kmeans.labels_))
```

Not so bad considering the algorithm is purely using the features to cluster the universities into 2 distinct groups. Hopefully you can begin to see how K Means is useful for clustering un-labeled data.
